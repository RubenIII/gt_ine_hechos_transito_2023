```python
from pyspark.sql.functions import *
from pyspark.sql.types import *
```

**LEER LOS DATOS DEL CONTENEDOR**

```python
# Constantes
str_container = "bronze"
str_storage_name = "stinetraficodeveastus2"
str_storage_path = f"abfss://{str_container}@{str_storage_name}.dfs.core.windows.net/"
str_file_lesionados = "07BaseDatosFallecidosLecionadosPNC2023.csv"
str_ruta_lesionados = str_storage_path + str_file_lesionados
```

```python
# Cargar cargar datos
df_lesionados_raw = spark.read.option('header', True)\
                        .option('inferSchema', True)\
                        .csv(str_ruta_lesionados)

# Quitar caracteres especiales nombres de columnas
df_lesionados_raw = df_lesionados_raw.withColumnRenamed("núm_corre", "num_corre")
df_lesionados_raw = df_lesionados_raw.withColumnRenamed("año_ocu", "anio_ocu")
df_lesionados_raw = df_lesionados_raw.withColumnRenamed("día_ocu", "dia_ocu")
df_lesionados_raw = df_lesionados_raw.withColumnRenamed("día_sem_ocu", "dia_sem_ocu")
df_lesionados_raw = df_lesionados_raw.withColumnRenamed("g_edad_80ymás", "g_edad_80ymas")
df_lesionados_raw = df_lesionados_raw.withColumnRenamed("g_edad_60ymás", "g_edad_60ymas")
```

**CATALOGOS CON ORDEN ALFANUMERICO**

```python
# Horas
df_horas = df_lesionados_raw.select('g_hora').distinct().orderBy('g_hora')
df_horas = df_horas.withColumn("id_g_hora", monotonically_increasing_id() + 1)
df_horas = df_horas.select("id_g_hora", "g_hora")
# Grupo de Horas
df_g_hora_5 = df_lesionados_raw.select('g_hora_5').distinct().orderBy('g_hora_5')
df_g_hora_5 = df_g_hora_5.withColumn("id_g_hora_5", monotonically_increasing_id() + 1)
df_g_hora_5 = df_g_hora_5.select("id_g_hora_5", "g_hora_5")
# Departamento Ocurrencia
df_depto_ocu = df_lesionados_raw.select('depto_ocu').distinct().orderBy('depto_ocu')
df_depto_ocu = df_depto_ocu.withColumn("id_depto_ocu", monotonically_increasing_id() + 1)
df_depto_ocu = df_depto_ocu.select("id_depto_ocu", "depto_ocu")
# Municipio Ocurrencia
df_mupio_ocu = df_lesionados_raw.select('mupio_ocu').distinct().orderBy('mupio_ocu')
df_mupio_ocu = df_mupio_ocu.withColumn("id_mupio_ocu", monotonically_increasing_id() + 1)
df_mupio_ocu = df_mupio_ocu.select("id_mupio_ocu", "mupio_ocu")
# Sexo
df_sexo_per = df_lesionados_raw.select('sexo_per').distinct().orderBy('sexo_per')
df_sexo_per = df_sexo_per.withColumn("id_sexo_per", monotonically_increasing_id() + 1)
df_sexo_per = df_sexo_per.select("id_sexo_per", "sexo_per")
# Mayor Menor Edad
df_mayor_menor = df_lesionados_raw.select('mayor_menor').distinct().orderBy('mayor_menor')
df_mayor_menor = df_mayor_menor.withColumn("id_mayor_menor", monotonically_increasing_id() + 1)
df_mayor_menor = df_mayor_menor.select("id_mayor_menor", "mayor_menor")
# Tipo Vehiculo
df_tipo_veh = df_lesionados_raw.select('tipo_veh').distinct().orderBy('tipo_veh')
df_tipo_veh = df_tipo_veh.withColumn("id_tipo_veh", monotonically_increasing_id() + 1)
df_tipo_veh = df_tipo_veh.select("id_tipo_veh", "tipo_veh")
# Marca Vehiculo
df_marca_veh = df_lesionados_raw.select('marca_veh').distinct().orderBy('marca_veh')
df_marca_veh = df_marca_veh.withColumn("id_marca_veh", monotonically_increasing_id() + 1)
df_marca_veh = df_marca_veh.select("id_marca_veh", "marca_veh")
# Color Vehiculo
df_color_veh = df_lesionados_raw.select('color_veh').distinct().orderBy('color_veh')
df_color_veh = df_color_veh.withColumn("id_color_veh", monotonically_increasing_id() + 1)
df_color_veh = df_color_veh.select("id_color_veh", "color_veh")
# Modelo Vehiculo
df_modelo_veh = df_lesionados_raw.select('modelo_veh').distinct().orderBy('modelo_veh')
df_modelo_veh = df_modelo_veh.withColumn("id_modelo_veh", monotonically_increasing_id() + 1)
df_modelo_veh = df_modelo_veh.select("id_modelo_veh", "modelo_veh")
# Grupo Modelo Vehiculo
df_g_modelo_veh = df_lesionados_raw.select('g_modelo_veh').distinct().orderBy('g_modelo_veh')
df_g_modelo_veh = df_g_modelo_veh.withColumn("id_g_modelo_veh", monotonically_increasing_id() + 1)
df_g_modelo_veh = df_g_modelo_veh.select("id_g_modelo_veh", "g_modelo_veh")
# Tipo Vehiculo
df_tipo_eve = df_lesionados_raw.select('tipo_eve').distinct().orderBy('tipo_eve')
df_tipo_eve = df_tipo_eve.withColumn("id_tipo_eve", monotonically_increasing_id() + 1)
df_tipo_eve = df_tipo_eve.select("id_tipo_eve", "tipo_eve")
# Fallecido o lesionado
df_fall_les = df_lesionados_raw.select('fall_les').distinct().orderBy('fall_les')
df_fall_les = df_fall_les.withColumn("id_fall_les", monotonically_increasing_id() + 1)
df_fall_les = df_fall_les.select("id_fall_les", "fall_les")
# Internado o No Internado
df_int_o_noint = df_lesionados_raw.select('int_o_noint').distinct().orderBy('int_o_noint')
df_int_o_noint = df_int_o_noint.withColumn("id_int_o_noint", monotonically_increasing_id() + 1)
df_int_o_noint = df_int_o_noint.select("id_int_o_noint", "int_o_noint")
```

**JOINS PARA CONVERTIR LOS CATALOGOS EN VALORES NUMERICOS**

```python
# Realizar los joins
df_lesionados_ingest = df_lesionados_raw \
    .join(df_horas, df_lesionados_raw.g_hora == df_horas.g_hora, "left") \
    .join(df_g_hora_5, df_lesionados_raw.g_hora_5 == df_g_hora_5.g_hora_5, "left") \
    .join(df_mes_ocu, df_lesionados_raw.mes_ocu == df_mes_ocu.mes_ocur, "left") \
    .join(df_dia_sem_ocu, df_lesionados_raw.dia_sem_ocu == df_dia_sem_ocu.mes_ocur, "left") \
    .join(df_depto_ocu, df_lesionados_raw.depto_ocu == df_depto_ocu.depto_ocu, "left") \
    .join(df_mupio_ocu, df_lesionados_raw.mupio_ocu == df_mupio_ocu.mupio_ocu, "left") \
    .join(df_sexo_per, df_lesionados_raw.sexo_per == df_sexo_per.sexo_per, "left") \
    .join(df_mayor_menor, df_lesionados_raw.mayor_menor == df_mayor_menor.mayor_menor, "left") \
    .join(df_tipo_veh, df_lesionados_raw.tipo_veh == df_tipo_veh.tipo_veh, "left") \
    .join(df_marca_veh, df_lesionados_raw.marca_veh == df_marca_veh.marca_veh, "left") \
    .join(df_color_veh, df_lesionados_raw.color_veh == df_color_veh.color_veh, "left") \
    .join(df_modelo_veh, df_lesionados_raw.modelo_veh == df_modelo_veh.modelo_veh, "left") \
    .join(df_g_modelo_veh, df_lesionados_raw.g_modelo_veh == df_g_modelo_veh.g_modelo_veh, "left") \
    .join(df_g_edad_80ymas, df_lesionados_raw.g_edad_80ymas == df_g_edad_80ymas.g_edad_80ymas, "left") \
    .join(df_g_edad_60ymas, df_lesionados_raw.g_edad_60ymas == df_g_edad_60ymas.g_edad_60ymas, "left") \
    .join(df_edad_quinquenales, df_lesionados_raw.edad_quinquenales == df_edad_quinquenales.edad_quinquenales, "left") \
    .join(df_tipo_eve, df_lesionados_raw.tipo_eve == df_tipo_eve.tipo_eve, "left") \
    .join(df_fall_les, df_lesionados_raw.fall_les == df_fall_les.fall_les, "left") \
    .join(df_int_o_noint, df_lesionados_raw.int_o_noint == df_int_o_noint.int_o_noint, "left")
# Seleccionar las columnas finales (IDs numéricos en lugar de cadenas)
df_lesionados_ingest = df_lesionados_ingest.select(
    "num_corre", 
    "anio_ocu", 
    "dia_ocu", 
    "hora_ocu",
    df_horas.id_g_hora.alias("id_g_hora"),
    df_g_hora_5.id_g_hora_5.alias("id_g_hora_5"),
    df_mes_ocu.id_mes_ocurrio.alias("id_mes_ocurrio"),
    df_dia_sem_ocu.id_dia_sem_ocu.alias("id_dia_sem_ocu"),
    df_depto_ocu.id_depto_ocu.alias("id_depto_ocu"),
    df_mupio_ocu.id_mupio_ocu.alias("id_mupio_ocu"),
    "zona_ocu",
    df_sexo_per.id_sexo_per.alias("id_sexo_per"),
    "edad_per",
    df_g_edad_80ymas.id_g_edad_80ymas.alias("id_g_edad_80ymas"),
    df_g_edad_60ymas.id_g_edad_60ymas.alias("id_g_edad_60ymas"),
    df_edad_quinquenales.id_edad_quinquenales.alias("id_edad_quinquenales"),
    df_mayor_menor.id_mayor_menor.alias("id_mayor_menor"),
    df_tipo_veh.id_tipo_veh.alias("id_tipo_veh"),
    df_marca_veh.id_marca_veh.alias("id_marca_veh"),
    df_color_veh.id_color_veh.alias("id_color_veh"),
    df_modelo_veh.id_modelo_veh.alias("id_modelo_veh"),
    df_g_modelo_veh.id_g_modelo_veh.alias("id_g_modelo_veh"),
    df_tipo_eve.id_tipo_eve.alias("id_tipo_eve"),
    df_fall_les.id_fall_les.alias("id_fall_les"),
    df_int_o_noint.id_int_o_noint.alias("id_int_o_noint"),
)
```
